```python
#Q1
import requests
from bs4 import BeautifulSoup

# Define the URL template
url_template = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{}"

num_pages = 50

product_urls = []
product_names = []
product_prices = []
ratings = []
review_counts = []

for page in range(1, num_pages + 1):
    # Construct the URL`
    url = url_template.format(page)
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find all product containers on the page
    product_containers = soup.find_all("div", {"data-component-type": "s-search-result"})
    
    # Extract information from each product container
    for container in product_containers:
        product_url = container.find("a", {"class": "a-link-normal s-no-outline"})["href"]
        product_urls.append("https://www.amazon.in" + product_url)
        
        product_name = container.find("span", {"class": "a-size-medium a-color-base a-text-normal"}).text
        product_names.append(product_name)
        
        product_price = container.find("span", {"class": "a-offscreen"}).text
        product_prices.append(product_price)
        
        rating = container.find("span", {"class": "a-icon-alt"}).text
        ratings.append(rating)
        
        review_count = container.find("span", {"class": "a-size-base"}).text
        review_counts.append(review_count)

for i in range(len(product_urls)):
    print("Product URL:", product_urls[i])
    print("Product Name:", product_names[i])
    print("Product Price:", product_prices[i])
    print("Rating:", ratings[i])
    print("Number of Reviews:", review_counts[i])
    print("--------------------")
```

```python
#Q2
import requests
from bs4 import BeautifulSoup
import csv

# Create a CSV file and define headers
csv_filename = "product_data.csv"
headers = ["Product URL", "Product Name", "Product Price", "Rating", "Number of Reviews", "Description", "ASIN", "Product Description", "Manufacturer"]

with open(csv_filename, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(headers)

# Define the URL template
url_template = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{}"

num_pages = 200

product_urls = []
product_names = []
product_prices = []
ratings = []
review_counts = []

for page in range(1, num_pages + 1):
    # Construct the URL
    url = url_template.format(page)
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find all product containers on the page
    product_containers = soup.find_all("div", {"data-component-type": "s-search-result"})
    
    # Extract information from each product container
    for container in product_containers:
        product_url = container.find("a", {"class": "a-link-normal s-no-outline"})["href"]
        product_urls.append("https://www.amazon.in" + product_url)
        
        product_name = container.find("span", {"class": "a-size-medium a-color-base a-text-normal"}).text
        product_names.append(product_name)
        
        product_price = container.find("span", {"class": "a-offscreen"}).text
        product_prices.append(product_price)
        
        rating = container.find("span", {"class": "a-icon-alt"}).text
        ratings.append(rating)
        
        review_count = container.find("span", {"class": "a-size-base"}).text
        review_counts.append(review_count)

        product_page_url = "https://www.amazon.in" + product_url
        product_response = requests.get(product_page_url)
        product_soup = BeautifulSoup(product_response.content, "html.parser")
    
        description_element = product_soup.find("div", {"id": "productDescription"})
        description = description_element.text.strip() if description_element else ""

        asin_element = product_soup.find("th", text="ASIN")
        asin = asin_element.find_next_sibling("td").text.strip() if asin_element else ""

        prod_element=product_soup.find("div", {"id": "feature-bullets"})
        product_description = prod_element.text.strip() if prod_element else ""
        
        manufac=product_soup.find("a", {"id": "bylineInfo"})
        manufacturer = manufac.text.strip() if manufac else ""

        with open(csv_filename, "a", newline="", encoding="utf-8") as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([product_url, product_name, product_price, rating, review_count, description, asin, product_description, manufacturer])

print("product_data.csv was created in the same directory")
```

```python
#Extra feature to sort the products that I hit by their rating
import csv

# Read the CSV file
csv_filename = "product_data.csv"

# Create a list to store modified and sorted data
modified_data = []

with open(csv_filename, "r", newline="", encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile)
    
    # Skip the header row
    headers = next(reader)
    
    for row in reader:
        # Extract the field containing "4.3 out of 5"
        rating_field = row[3]  # Assuming rating is at index 3
        
        # Extract the numeric part before " out of 5"
        rating_value = float(rating_field.split(" out of 5")[0])
        
        # Modify the row by replacing the rating field with the extracted value
        row[3] = rating_value
        
        # Append the modified row to the list
        modified_data.append(row)

# Sort the modified data based on the rating (ascending order)
sorted_data = sorted(modified_data, key=lambda x: x[3], reverse=True)

# Write the sorted data back to the CSV file
with open(csv_filename, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    
    # Write the headers
    writer.writerow(headers)
    
    # Write the sorted data
    writer.writerows(sorted_data)
```
